```python
numeric_data = data.select_dtypes(include=['int64', 'float64']).columns
non_numeric_data = data.select_dtypes(exclude=['int64', 'float64']).columns
print(f"Columns of type int or float: {len(numeric_data)}, Columns with other type: {len(non_numeric_data)}")
```

```python
#Investigate the number of unique values in each column sorted by datatype
unique_values_summary=[]
for column in data.columns:
    unique_values_summary.append(len(data[column].unique()))
column_name=data.columns
titles = ['Variable', 'Nb_unique_values']
dtypes_list = data.dtypes.tolist()
summary=pd.DataFrame({
    'Variable': column_name,
    'Number_of_unique_values': unique_values_summary,
    'Datatype':dtypes_list
})
summary
```
